#### 4. 查看知识图谱相关表格

````python
import pandas as pd
from tabulate import tabulate

# 正确的设置 - 允许换行但保持格式整齐
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 120)  # 设置合适的宽度，允许自然换行
pd.set_option('display.max_colwidth', 50)  # 控制单行最大宽度，促进换行

# 设置左对齐
pd.set_option('display.colheader_justify', 'right')  # 列标题左对齐
pd.set_option('display.unicode.east_asian_width', True)  # 处理中文字符对齐
````
